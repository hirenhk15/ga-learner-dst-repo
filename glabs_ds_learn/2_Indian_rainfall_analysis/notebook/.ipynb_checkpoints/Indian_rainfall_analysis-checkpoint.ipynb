# Indian Rainfall Analysis

The 2019 Indian floods were a series of floods that affected over thirteen states in late July and early August 2019, due to incessant rains. At least 200 people died and about a million people were displaced. Karnataka and Maharashtra were the most severely affected states.

It was the heaviest monsoon in the last 25 years. More than 1600 people died between June and October 2019.

India being an agriculturally driven economy, it will be interesting to study the rainfall in India in the past decade to give us an idea of the changes in the pattern if there are any.

Let us work on the INDIAN RAINFALL DATA!

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read the data

```python
data = pd.read_csv('../data/rainfall in india 1901-2015.csv')
data.head()
```

## Summary of the data

```python
data.info()
```

## Inspect the data, find missing values and replace them with appropriate values

```python
data.isnull().sum()
```

```python
data.mean()
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.isnull().sum()
```

## Plot the mean annual rainfall and note down your observations regarding the same

```python
# Calculate mean average rainfall and moving average of 10 years
mean_annual_rainfall = data.groupby('YEAR')['ANNUAL'].mean()
MA10 = mean_annual_rainfall.rolling(10).mean()

plt.figure(figsize=(12, 7))
mean_annual_rainfall.plot(color='b', marker='o', linewidth=2, figsize=(12,7), label='Mean Average Rainfall')
MA10.plot(color='r', linewidth=2, label='Moving Average of 10 years')

plt.xlabel('Year', fontsize=18)
plt.ylabel('Annual Rainfall (in mm)', fontsize=18)
plt.title('Annual Rainfall in India from Year 1901 to 2015', fontsize=22)
plt.tick_params(labelsize=12)
plt.legend()
plt.grid()
plt.show()
```

1. Highest average rainfall in India was recored in the year 1961. This was because in 1961 India received multiple cyclones.City of Pune was flooded in the year 1961 which is remembered as Panshet Flood.

2. Year 1965-66 were twin drought years and there was food scarcity in India. Prime Minister Lal Bahadur Shastri gave the Slogan Jai Jawan Jai Kissan to people of India. This lead to green revolution in India making India a food surplus country in the coming decades.

3. The red line is the 10 year moving average of the rainfall in India. It seems since 1960s there is slight dip in the rainfall in India. Now a days due to global warming the period of Monsoon season has shortned. We see more of erratic rainfall pattern. This needs more study.

## Similarly analyze the seasonal rainfall as per subdivisions and note down your observations

```python
seasonal_rainfall = data.groupby('SUBDIVISION')[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']].mean()
seasonal_rainfall.sort_values('Jun-Sep').plot.barh(width=0.5, edgecolor='k', align='center', stacked=True, figsize=(12,10))
plt.xlabel('Rainfall (in mm)', fontsize=18)
plt.ylabel('Subdivision', fontsize=18)
plt.title('Seasonal Rainfall in Subdivisions of India', fontsize=20)
plt.show()
```

1. From the above graph we can see that majority of rainfall is received in the month of Jun-Sep which is the Monsoon season. Oct-Dec is time of return monsoon. Jan-Feb are the winter months.Mar-May is time for Summer rains.

2. Coastal Karnataka, Arunachal Pradesh, Konkan Goa and Kerala receive highest rainfall.

3. Rajastan, Haryana and Punjab receives low rainfall. Interesting thing is that Punjab and Haryana have high agricultural output despite low rainfall.Their water requirnments are met by rivers and canals.

## With boxplot analyze the distribution of rainfall in various states and note down your observations

```python
plt.figure(figsize=(12,15))
ax = sns.boxplot(x='ANNUAL', y='SUBDIVISION', data=data)
ax.set_xlabel('Subdivision', fontsize=15)
ax.set_ylabel('Annual Rainfall (in mm)', fontsize=15)
plt.title('Annual Rainfall in Subdivisions of India', fontsize=22)
plt.grid()
plt.show()
```

1. We can see Subdivision Arunachal Pradesh shows highest difference between Maximum and Minimum rainfall received. Costal Karnataka receives close to 3400 mm of Annual rainfall which is the highest in India. West Rajastan receives the least amount of rainfall.

## Analyze the rainfall through years in Kerala and note down your observations

```python
kerala_annual = data[data['SUBDIVISION']=='KERALA'].groupby('YEAR')['ANNUAL'].mean()

plt.figure(figsize=(12, 7))
kerala_annual.plot(color='b', marker='o', linewidth=2, figsize=(12,7), label='Mean Average Rainfall in Kerala')
mean_annual_rainfall.plot(color='r', linewidth=2, label='Mean Average Rainfall in India')

plt.xlabel('Year', fontsize=18)
plt.ylabel('Kerala Annual Rainfall (in mm)', fontsize=18)
plt.title('Annual Rainfall in Kerala from Year 1901 to 2015', fontsize=22)
plt.tick_params(labelsize=12)
plt.legend()
plt.grid()
plt.show()
```

Prior to 2018 Kerala had major flood in the year 1924 which is evident in the data. Contrary to popular belief Kerala received maximum annual rainfall in year 1961(4257 mm) and not 1924(4226 mm). In 2018 Kerala has received 2226.4 mm of rain in the monsoon season. This is 40% more than the average rainfall.

## Read the district wise rainfall data

```python
dist = pd.read_csv('../data/district wise rainfall normal.csv')
dist.head()
```

## Annual rainfall in different districts of Kerala

```python
kerala_dist = dist[dist['STATE_UT_NAME']=='KERALA'].groupby('DISTRICT')['ANNUAL'].mean()

kerala_dist.sort_values().plot.barh(edgecolor='k', align='center', linewidth=2, figsize=(10, 8))
plt.xlabel('District', fontsize=18)
plt.ylabel('Annual Rainfall (in mm)', fontsize=18)
plt.title('Rainfall in Districts of Kerala', fontsize=22)
plt.tick_params(labelsize=12)
plt.show()
```

## Find out the districts with least rainfall

```python
least_rainfall_dist = dist.groupby(['STATE_UT_NAME', 'DISTRICT'])['ANNUAL'].min().nsmallest(10)

least_rainfall_dist.plot.barh(width=0.5, edgecolor='k', align='center', linewidth=2, figsize=(10,8))
plt.xlabel('Annual Rainfall (in mm)', fontsize=18)
plt.ylabel('District', fontsize=18)
plt.title('Top 10 Districts With Minumum Rainfall in India', fontsize=22)
plt.tick_params(labelsize=12)
plt.show()
```

Ladak which is part of Jammu and Kashmir receives 94.6 mm. Ladak and Kargil which receive less rainfall are part of Indian State Jammu and Kashmir. Jaisalmer, Sri Ganganaga and Barmer are part of Rajastan State.

## Similarly, find districts with maximum rainfall

```python
max_rainfall_dist = dist.groupby(['STATE_UT_NAME', 'DISTRICT'])['ANNUAL'].max().nlargest(10)

max_rainfall_dist.sort_values().plot.barh(width=0.5, edgecolor='k', align='center', linewidth=2, figsize=(10,8))
plt.xlabel('Annual Rainfall (in mm)', fontsize=18)
plt.ylabel('District', fontsize=18)
plt.title('Top 10 Districts With Maximum Rainfall in India', fontsize=22)
plt.tick_params(labelsize=12)
plt.show()
```

## Crop data

```python
crop = pd.read_csv('../data/crop_data.csv')
crop.head()
```

```python
# Production of crops per season
crop.groupby('Season')['Production'].mean().plot.barh(edgecolor='k', linewidth=2, figsize=(12, 7))
plt.ylabel('Season', fontsize=18)
plt.xlabel('Production', fontsize=18)
plt.title('Production Of Crops Per Season', fontsize=22)
plt.tick_params(labelsize=12)
plt.show()
```

```python
# Production of crops per year 
crop.groupby('Crop_Year')['Production'].mean().plot.barh(edgecolor='k', linewidth=2, figsize=(12, 7))
plt.ylabel('Crop Year', fontsize=18)
plt.xlabel('Production', fontsize=18)
plt.title('Production Of Crops Per Year', fontsize=22)
plt.tick_params(labelsize=12)
plt.show()
```

```python
annual_prod = crop.groupby('Crop_Year')['Production'].mean()
annual_prod.index
```

```python
mean_annual_rainfall.index
```

```python
crop_rainfall = pd.concat([annual_prod, mean_annual_rainfall], axis=1, join='outer').dropna()

# Relation between annual rainfall and crop production
plt.figure(figsize=(15, 7))
plt.scatter(crop_rainfall['Production'], crop_rainfall['ANNUAL'])
plt.xlabel('Crop Production', fontsize=18)
plt.ylabel('Mean Annual Rainfall', fontsize=18)
plt.title('Crop Production w.r.t. Mean Annual Rainfall between 1997-2015', fontsize=22)
plt.show()
```

It seems there's no as such correlation between crop production and annual rainfall.
